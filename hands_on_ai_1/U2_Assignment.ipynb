        <li>Evaluate which two words have a higher similarity: "study" and "exam" or "party" and "exam".</li>

```python
study = word_embedding_english('study')
party = word_embedding_english('party')
exam = word_embedding_english('exam')

party_exam = party.similarity(exam)
study_exam = study.similarity(exam)

if study_exam > party_exam:
    print(f'"study" and "exam" have a higher similarity, equal to {study_exam}')
else:
    print(f'"party" and "exam" have a higher similarity, equal to {party_exam}')
```
